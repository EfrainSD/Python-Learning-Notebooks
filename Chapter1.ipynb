* Syntax Error: It will not allow any of your code to run due to a syntax error. For example: print(1
* Runtime Error: It does not appear until you run the program, and usually occurs when something unusual has happened. For example: print(1 + “Hello”)
* Semantic Error: These are the most difficult to catch. They will let the program run and will not indicate that anything is wrong. For example, if you want to add two numbers and enter print(“1” + “5”), the result will be 15, which is not wrong and does not cause an error, but it does not do what we want.
