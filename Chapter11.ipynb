<h1>Lists are Mutable</h1>

```python
my_list = ["T", "E", "S", "T"]
my_list[2] = "X"
my_list
```

```python
a_list = ["a", "b", "c", "d", "e", "f"]
a_list[1:3] = ["x", "y"]
a_list
```

<h1> Elements can be deleted </h1>

```python
a_list = ["a", "b", "c", "d", "e", "f"]
del a_list[1:5]
a_list
```

<h1> Objects and References </h1>

```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)
print(a is b)
```

If we equate one list to another, both point to the same object.

```python
a = [1, 2, 3]
b = a
print(a is b)

b[0] = 5
print(a)
```

To clone it, we use [:]

```python
a = [1, 2, 3]
b = a[:]
b
```

<h1> List methods </h1>

Add

```python
mylist = []
mylist.append(5)
mylist.append(3)
mylist.append(12)
mylist
```

Insert in a position

```python
mylist.insert(1, 9)
mylist
```

Count how many times a value appears

```python
mylist.count(12)
```

Put a list at the end.

```python
mylist.extend([5, 9, 5, 11]) # Put whole list onto end of mylist
mylist
```

Find the index of the first 12 in the list

```python
mylist.index(12)
```

Flips the list

```python
mylist.reverse()
mylist
```

Sort the list

```python
mylist.sort()
mylist
```

Remove the first 12 from the list

```python
mylist.remove(12)
mylist
```
